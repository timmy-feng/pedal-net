```python
%pip install matplotlib
import matplotlib.pyplot as plt
import numpy as np

# Example data
categories_old = [
    "Naive F1 Score 69", "Naive F1 Score 71", "Naive F1 Score 72", 
    "Ground F1 Score 69", "Ground F1 Score 71", "Ground F1 Score 72"
]
means_old = [0.4248274915, 0.4029099501, 0.3879176307, 0.0357389449, 0.04288046242, 0.04047623393]
conf_intervals_old = [0.05378900195, 0.04747977244, 0.05350087564, 0.009618050485, 0.01171648491, 0.01375501807]  # Confidence intervals

categories = [categories_old[i] for i in [0, 3, 1, 4, 2, 5]]
means = [means_old[i] for i in [0, 3, 1, 4, 2, 5]]
conf_intervals = [conf_intervals_old[i] for i in [0, 3, 1, 4, 2, 5]]

# Extract labels (e.g., 69, 71, 72) from category names
labels = [label.split()[-1] for label in categories]

# Assign colors based on the labels
colors = ['red' if 'Naive' in label else 'blue' for label in categories]

# Plot
x_pos = np.arange(len(categories))
plt.bar(x_pos, means, yerr=conf_intervals, capsize=5, color=colors, edgecolor='black')

# Customize
plt.xticks(x_pos, categories, rotation=45, ha='right')  # Rotate labels for clarity
plt.ylabel('F1 Score')
plt.title('F1 Score for 95% Confidence Intervals')

# Show
plt.tight_layout()  # Adjust layout to prevent overlap
plt.show()
```
